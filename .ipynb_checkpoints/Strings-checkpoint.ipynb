## Indexing and Slicing

### Indexing

```python
word = 'hire'
```

```python
print(word)
```

```python
word[0] #Indexing
```

```python
word[-1] # Negetive Index
```

### Slicing

```python
anotherword = 'abcdef'
```

```python
anotherword[0:3] #strart at 0 go upto 3rd character (but exclude 3rd)
```

```python
anotherword[2:2] #start at 2 and upto 2 means no characters are grabbed
```

Get everything till end from a starting position

```python
anotherword[2:] #start at 2 and grab everything till end
```

```python
anotherword[:2] #start at beginning upto 2
```

### Slicing with a jump (Step sizes)

```python
anotherword[1:6:2] #start position: stop position: step size
```

## String Built In Methods

```python
basic = 'hello world'
```

```python
basic.upper() #Returns basic win upper case. But does not change basic. Basic is still smallcase 'hello world'
```

```python
basic #The variable basic has not changed.
```

```python
basic.split() #Splits a senence with spaces. Returns a list
```

```python
basic.split('o')
```

### String interpolation

```python
user_name = 'Recruit'
action = 'run'

print('The {a} needs to {b}'.format(a=user_name, b=action))
```

### String Interpolation for numbers

```python
num = 123.6789
print('The code is {:.1F}'.format(num)) #format as 1 decimal place
```

```python
print('The code is {:.2F}'.format(num)) #format as 2 decimal place
```

```python
print('The code is {:.0F}'.format(num)) #format as a whole number
```
